```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
```

```python
# File to Load
JC_2016_tripdata = Path("Resources/JC-201606-citibike-tripdata.csv")
JC_2021_tripdata = Path("Resources/JC-202101-citibike-tripdata.csv")

# Read JC Data File and store into Pandas DataFrames
JC_2016_df = pd.read_csv(JC_2016_tripdata)
JC_2O21_df = pd.read_csv(JC_2021_tripdata)
```

```python
# Display the data table for JC_2016
JC_2016_df
```

```python
# Display the data table for JC_2O21
JC_2O21_df
```

```python
# Rename columns for JC_2021
JC_2021_renamed_df = JC_2O21_df.rename(columns={"tripduration": "Trip Duration", "starttime": "Start Time",
                                                "stoptime": "Stop Time", "start station id": "Start Station ID",
                                               "start station name": "Start Station Name",
                                                "start station latitude": "Start Station Latitude", 
                                               "start station longitude": "Start Station Longitude",
                                               "end station id": "End Station ID",
                                                "end station name": "End Station Name",
                                               "end station latitude": "End Station Latitude",
                                               "end station longitude": "End Station Longitude",
                                               "bikeid": "Bike ID", "usertype": "User Type",
                                                "birth year": "Birth Year", "gender": "Gender"})
JC_2021_renamed_df
```

```python
# Combine DataFrames
JC_combined_df = pd.concat([JC_2016_df, JC_2021_renamed_df], axis=0, ignore_index=True)
JC_combined_df
```

```python
# Drop null values
JC_combined_cleaned_df = JC_combined_df.dropna(how="any")
JC_combined_cleaned_df
```

```python
# Drop rows with 0 for Gender
JC_cleaned_tripdata_df = JC_combined_cleaned_df.drop(JC_combined_cleaned_df[JC_combined_cleaned_df["Gender"] ==0].index)
JC_cleaned_tripdata_df
```

```python
# Identify data type
JC_cleaned_tripdata_df.info()
```

```python
# Change birth year data type to int64
JC_cleaned_tripdata_df["Birth Year"] = JC_cleaned_tripdata_df["Birth Year"].astype("int64")
JC_cleaned_tripdata_df
```

```python
# Identify the user type
JC_cleaned_tripdata_user_df = JC_cleaned_tripdata_df["User Type"].unique().tolist()
JC_cleaned_tripdata_user_df
```

```python
# Identify the gender
JC_cleaned_tripdata_gender_df = JC_cleaned_tripdata_df["Gender"].unique().tolist()
JC_cleaned_tripdata_gender_df
```

```python
# Add column Trip ID and set it as index
# Generate a list of index values (e.g., 0, 1, 2)
new_index_values = list(range(len(JC_cleaned_tripdata_df)))

# Assign the list as the index to the DataFrame
JC_cleaned_tripdata_df.index = new_index_values
JC_cleaned_tripdata_df["Trip ID"] = new_index_values
JC_cleaned_tripdata_df.set_index("Trip ID", inplace=True)
```

```python
# Review the dataframe
JC_cleaned_tripdata_df
```

```python
# Save combined, cleaned csv file
# Exporting JC_cleaned_tripdata_df into csv file
JC_cleaned_tripdata_df.to_csv("Resources/JC_cleaned_tripdata.csv", index = True, header = True)
```
